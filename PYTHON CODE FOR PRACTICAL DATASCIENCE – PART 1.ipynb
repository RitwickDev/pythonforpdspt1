```python
#Task 1: Data Preparation
# Importing pandas
import pandas as pd

# Storing as a csv object
auto = 'automobile.csv'

# Reading csv file
auto = pd.read_csv(auto, sep='#', decimal='.', header=None, names= ['symboling','normalized-losses', 'make',
'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels','engine-location', 'wheel-base', 'length', 'width',
'height', 'curb-weight', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio',
'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price'])

# Displaying top few instances the file
auto.head()

# Displaying data types for all the columns
auto.dtypes

# Checking the Data structured 
auto.shape

# Checking different values with count of records for each value.

# Symboling of the car
print(auto['symboling'].value_counts())
# Checking minimum and maximum value.
min(auto['symboling'])
max(auto['symboling'])
# Specifing range and taking values only present in the range.
auto['symboling'] = auto['symboling'][auto['symboling'].isin([-3, -2, -1, 0, 1, 2, 3])]
# Checking all values are again
print(auto['symboling'].value_counts())

# Normalized-losses of the car
print(auto['normalized-losses'].value_counts())
min(auto['normalized-losses'])
max(auto['normalized-losses'])
# Values ranging from 65 to 256 taken into consideration
auto['normalized-losses'] = auto['normalized-losses'][auto['normalized-losses'].isin(range(65, 257))]
# Plotting histogram for skewness
print(auto['normalized-losses'].plot(kind='hist', bins = 10))
# Imputing based on median
auto['normalized-losses'] = auto['normalized-losses'].fillna(auto['normalized-losses'].median(axis=0))
print(auto['normalized-losses'].value_counts())

#Make of the Car
print(auto['make'].value_counts())
# Romving whitespaces 
auto['make']=auto['make'].str.strip()
# Converting all values to lower case
auto['make'] = auto['make'].str.lower()
# Replacing typo
auto['make'] = auto['make'].replace('vol00112ov', 'volvo')
print(auto['make'].value_counts())

# Performing the same data preprocessing steps of removing whitespaces, replacing typos converting into lower case
# and treating for NA values.

# Fuel Type of the car
print(auto['fuel-type'].value_counts())
auto['fuel-type'] = auto['fuel-type'].str.strip()
auto['fuel-type'] = auto['fuel-type'].str.lower()
print(auto['fuel-type'].value_counts())

# Aspiration of the car
print(auto['aspiration'].value_counts())
auto['aspiration'] = auto['aspiration'].str.strip()
auto['aspiration'] = auto['aspiration'].str.lower()
auto['aspiration'] = auto['aspiration'].replace('turrrrbo', 'turbo')
print(auto['aspiration'].value_counts())

# Number of Doors of the car
print(auto['num-of-doors'].value_counts())
auto['num-of-doors'] = auto['num-of-doors'].str.strip()
auto['num-of-doors'] = auto['num-of-doors'].str.lower()
auto['num-of-doors'] = auto['num-of-doors'].replace('fourr', 'four')
# Dropping all null values
auto['num-of-doors'] = auto['num-of-doors'].dropna()
print(auto['num-of-doors'].value_counts())

#Body style of the car
print(auto['body-style'].value_counts())
auto['body-style'] = auto['body-style'].str.strip()
auto['body-style'] = auto['body-style'].str.lower()
print(auto['body-style'].value_counts())

# Wheel drive of the car
print(auto['drive-wheels'].value_counts())
auto['drive-wheels'] = auto['drive-wheels'].str.strip()
auto['drive-wheels'] = auto['drive-wheels'].str.lower()
print(auto['drive-wheels'].value_counts())

# Engine Location of the car
print(auto['engine-location'].value_counts())
auto['engine-location'] = auto['engine-location'].str.strip()
auto['engine-location'] = auto['engine-location'].str.lower()
print(auto['engine-location'].value_counts())

# Wheel Base of the car
print(auto['wheel-base'].value_counts())
min(auto['wheel-base'])
max(auto['wheel-base'])
print(auto['wheel-base'].value_counts())

# Length of the car
print(auto['length'].value_counts())
min(auto['length'])
max(auto['length'])
print(auto['length'].value_counts())

# Width of the car
print(auto['width'].value_counts())
min(auto['width'])
max(auto['width'])
print(auto['width'].value_counts())

# Height of the car
print(auto['height'].value_counts())
min(auto['height'])
max(auto['height'])
print(auto['height'].value_counts())

# Curb weight of the car
print(auto['curb-weight'].value_counts())
min(auto['curb-weight'])
max(auto['curb-weight'])
print(auto['curb-weight'].value_counts())

# Engine type of the car
print(auto['engine-type'].value_counts())
auto['engine-type'] = auto['engine-type'].str.strip()
auto['engine-type'] = auto['engine-type'].str.lower()
print(auto['engine-type'].value_counts())

# Number of cylinder in the car
print(auto['num-of-cylinders'].value_counts())
auto['num-of-cylinders'] = auto['num-of-cylinders'].str.strip()
auto['num-of-cylinders'] = auto['num-of-cylinders'].str.lower()
print(auto['num-of-cylinders'].value_counts())

# Engine size of the car
print(auto['engine-size'].value_counts())
min(auto['engine-size'])
max(auto['engine-size'])
print(auto['engine-size'].value_counts())

# Fuel System of the car
print(auto['fuel-system'].value_counts())
auto['fuel-system'] = auto['fuel-system'].str.strip()
auto['fuel-system'] = auto['fuel-system'].str.lower()
print(auto['fuel-system'].value_counts())

# Bore of the car
print(auto['bore'].value_counts())
min(auto['bore'])
max(auto['bore'])
auto['bore'] = auto['bore'].dropna()
print(auto['bore'].value_counts())

# Stroke of the car
print(auto['stroke'].value_counts())
min(auto['stroke'])
max(auto['stroke'])
auto['stroke'] = auto['stroke'].dropna()
print(auto['stroke'].value_counts())

# Compression Ratio of the car
print(auto['compression-ratio'].value_counts())
min(auto['compression-ratio'])
max(auto['compression-ratio'])
print(auto['compression-ratio'].value_counts())

# Horsepower of the car
print(auto['horsepower'].value_counts())
min(auto['horsepower'])
max(auto['horsepower'])
auto['horsepower'] = auto['horsepower'].dropna()
print(auto['horsepower'].value_counts())

# Peak-rpm of the car
print(auto['peak-rpm'].value_counts())
min(auto['peak-rpm'])
max(auto['peak-rpm'])
auto['peak-rpm'] = auto['peak-rpm'].dropna()
print(auto['peak-rpm'].value_counts())

# City mileage of the car
print(auto['city-mpg'].value_counts())
min(auto['city-mpg'])
max(auto['city-mpg'])
print(auto['city-mpg'].value_counts())

# Highway mileage of the car
print(auto['highway-mpg'].value_counts())
min(auto['highway-mpg'])
max(auto['highway-mpg'])
print(auto['highway-mpg'].value_counts())

# Price of the car
print(auto['price'].value_counts())
min(auto['price'])
max(auto['price'])
auto['price'] = auto['price'][auto['price'].isin(range(5118, 45401))]
auto['price'] = auto['price'].dropna()
print(auto['price'].value_counts())
```

```python
# Task 2: Data Exploration
# Nominal values
# Numerical values

# Importing matplotlib.pyplot 
import matplotlib.pyplot as plt

# pie chart for fuel-type
auto['fuel-type'].value_counts().plot(kind='pie',autopct='%.3f',figsize = (7,7), title='Pie Chart for fuel-type in the car')

# pie chart for num-of-cylinders
auto['num-of-cylinders'].value_counts().plot(kind='pie',autopct='%.2f',figsize = (12,12), title='Pie Chart for number of Cylinders in the car')
# histogram for price
auto['price'].plot(kind='hist', bins = 10,figsize = (5,5),title='Histogram for price')
# changing default label and adding USD
plt.xlabel("price (USD)")
```

```python
# Task 2: Data Exploration
# 2. Explore the relationships between columns; at least 3 visualisations with plausible hypothesis

# Boxplot for City mileage and fuel-type
auto.boxplot(column='price', by='fuel-type', figsize = (5,5))
plt.xlabel("City miles per gallon")
plt.ylabel("Price in (USD)")

# Boxplot for price and number of cylinders
auto.boxplot(column='price', by='num-of-cylinders', figsize = (5,5))
plt.xlabel("Number of cylinders")
plt.ylabel("Price in (USD)")

# Scatter plot for engine size and price
auto.plot(kind= 'scatter', x=16, y=25, figsize = (5,5))
plt.title("Scatter plot for Engine size and Price")
plt.ylabel("Price in (USD)")
plt.xlabel("Engine Size")
```
